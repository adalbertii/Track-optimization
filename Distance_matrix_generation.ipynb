```python
import csv
import random

def generate_sample_csv(file_path, num_nodes=100):
    """
    Generates a symmetric distance matrix and saves it as a CSV file.
    """
    matrix = []
    for i in range(num_nodes):
        row = []
        for j in range(num_nodes):
            if i == j:
                row.append(0)
            elif j < i:
                # To ensure symmetry, copy from earlier row
                row.append(matrix[j][i])
            else:
                dist = random.randint(1, 100)
                row.append(dist)
        matrix.append(row)

    # Save to CSV
    with open(file_path, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        for row in matrix:
            writer.writerow(row)

# Generate the CSV
generate_sample_csv('distance_matrix.csv')
print("Sample distance matrix CSV generated.")
```
